```python
import hashlib

def get_md5(file_path: str) -> str:
    """Calcula MD5 de un archivo."""
    with open(file_path, "rb") as f:
        return hashlib.md5(f.read()).hexdigest()

# Uso
print(get_md5(r"persistence\files\landing\SBSA\VISA_Inward_Settlement_to_SBSA_BASEII_20251118.TXT").upper())
```
