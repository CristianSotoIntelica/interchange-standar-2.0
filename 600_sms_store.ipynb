```python
from interchange.persistence.file import FileStorage
from interchange.visa import transform, extract, clean, calculate, interchange, store

layer = FileStorage.Layer
```

```python
client_id = "SBSA"
file_id = "B6781ADDCFE0CD800BFA2968A6ED2816"
```

```python
origin_layer = layer.STAGING
target_layer = layer.OPERATIONAL
client_id = client_id
file_id = file_id
transactions_subdir = "300-SMS_CLN_MESSAGES"
calculated_subdir = "400-SMS_CAL_MESSAGES"
interchange_subdir = "500-SMS_ITX_MESSAGES"
target_subdir = "SMS_MESSAGES"
```

```python
print(
    origin_layer,
    target_layer,
    client_id,
    file_id,
    transactions_subdir,
    calculated_subdir,
    interchange_subdir,
    target_subdir,
)
```

```python
from interchange.logs.logger import Logger
from interchange.persistence.file import FileStorage


log = Logger(__name__)
fs = FileStorage()
```

```python
transactions = fs.read_parquet(
    origin_layer,
    client_id,
    file_id,
    subdir=transactions_subdir,
)
```

```python
transactions
```

```python
calculated = fs.read_parquet(
    origin_layer,
    client_id,
    file_id,
    subdir=calculated_subdir,
)
```

```python
calculated
```

```python
interchange = fs.read_parquet(
    origin_layer,
    client_id,
    file_id,
    subdir=interchange_subdir,
)
```

```python
merged_data = transactions.join(calculated, how="left", lsuffix="_sms")
merged_data = merged_data.join(interchange, how="left", rsuffix="_intelica")
```

```python
merged_data
```

```python
fs.write_parquet(merged_data, target_layer, client_id, file_id, subdir=target_subdir)
```
